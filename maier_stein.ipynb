```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Force field F(x,y) = [x-x**3 - alpha*x*y**2, -my*y*(1+x**2)]
# a(x,y) = x-x**3 - alpha*x*y**2
# b(x,y) = -my*y*(1+x**2)
```

```python
def a(x,y,alpha):
    return x-x**3 - alpha*x*y**2

def b(x,y,my):
    return -my*y*(1+x**2)



def F(x,y,alpha,my):
    return np.array([a(x,y,alpha), b(x,y,my)])
```

```python
alpha = 10
mu = 0.67
x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)
X, Y = np.meshgrid(x, y)
# U = a(X,Y,alpha)
# V = b(X,Y,my)

U, V = F(X, Y, alpha, mu)
plt.streamplot(X, Y, U, V, density=2, linewidth=0.5, arrowsize=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase space')
plt.show()

# Steady states at (-1,0) and (1,0)
# separatix at y = 0 and x = 0
```

```python
# Potential energy surface plot
def V(x,y,alpha,my):
    return 0.5*x**2 + 0.25*x**4 + 0.5*my*y**2 + 0.5*alpha*x**2*y**2

Z = V(X,Y,alpha,mu)
plt.contour(X, Y, Z, 500)
plt.xlabel('x')
plt.ylabel('y')

# Density colour label on the side
plt.colorbar()
```
