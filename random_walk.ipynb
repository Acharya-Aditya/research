```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# 1D random walk using gillespie algorithm

X = [0]
t = [0]

t_end = 100000

left = 0.5 # probability of moving left
right = 0.5 # probability of moving right

while t[-1] < t_end:

    tau = np.random.exponential(1/(left + right))
    t.append(t[-1] + tau)

    current_pos = X[-1]

    rand_prob = random.uniform(0,1)
    if rand_prob < left:
        new_pos = current_pos - 1
        X.append(new_pos)
    else:
        new_pos = current_pos + 1
        X.append(new_pos)
```

```python
plt.plot(t,X)
```

```python
# 2D random walk

X = [0]
Y = [0]

t = [0]

t_end = 1000

left = 0.25 # probability of moving left
right = 0.25 # probability of moving right
up = 0.25 # probability of moving up
down = 0.25 # probability of moving down

while t[-1] < t_end:

    tau = np.random.exponential(1/(left + right + up + down))
    t.append(t[-1] + tau)

    current_pos_x = X[-1]
    current_pos_y = Y[-1]

    rand_prob = random.uniform(0,1)
    if rand_prob < left:
        new_pos_x = current_pos_x - 1
        new_pos_y = current_pos_y
    elif rand_prob < left + right:
        new_pos_x = current_pos_x + 1
        new_pos_y = current_pos_y
    elif rand_prob < left + right + up:
        new_pos_x = current_pos_x
        new_pos_y = current_pos_y + 1
    else:
        new_pos_x = current_pos_x
        new_pos_y = current_pos_y - 1

    X.append(new_pos_x)
    Y.append(new_pos_y)

```

```python
plt.plot(Y,X)
```
