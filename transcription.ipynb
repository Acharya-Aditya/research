```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
X = [0]
t = [0]
# gonna plot X vs t

t_end = 10000

k = 2 # production rate
gamma = 0.1 # decay rate

while t[-1] < t_end:
    current_X = X[-1]

    rates = [k, gamma*current_X]
    rate_sum = sum(rates)

    tau = np.random.exponential(scale=1/rate_sum)

    t.append(t[-1] + tau)

    rand_prob = random.uniform(0,1)

    if rand_prob*rate_sum > 0 and rand_prob*rate_sum <rates[0]:
        X.append(X[-1] + 1)

    else:
        X.append(X[-1] -1)
```

```python
plt.figure(figsize=(8,8))
plt.plot(t,X)
```
